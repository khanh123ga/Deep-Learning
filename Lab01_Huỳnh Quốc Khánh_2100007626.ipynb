Bài 4:Ngân hàng bạn đang làm có chương trình cho vay ưu đãi cho các đối tượng mua
chung cư. Tuy nhiên gần đây có một vài chung cư rất hấp dẫn (giá tốt, vị trí đẹp,...) nên lượng hồ sơ người nộp cho chương trình ưu đãi tăng đáng kể. Bình
thường bạn có thể duyệt 10-20 hồ sơ một ngày để quyết định hồ sơ có được cho
vay hay không, tuy nhiên gần đây bạn nhận được 1000-2000 hồ sơ mỗi ngày.
Bạn không thể xử lý hết hồ sơ và bạn cần có một giải pháp để có thể dự đoán hồ
sơ mới là có nên cho vay hay không. Sử dụng hàm Sigmoid.


```python
from google.colab import drive
drive.mount('/content/gdrive')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

path='/content/gdrive/MyDrive/Dataset/dataset.csv'

data=pd.read_csv(path).values
N,d=data.shape
x=data[: ,0:d-1].reshape(-1,d-1)
y=data[: , 2].reshape(-1,1)

cho_vay=x[y[:, 0]==1]
tu_choi=x[y[:, 0]==0]

plt.scatter(cho_vay[:,0],cho_vay[:,1],c='r',edgecolors='none',s=30,label='Cho vay')
plt.scatter(tu_choi[:,0],tu_choi[:,1],c='b',edgecolors='none',s=30,label='Từ chối')
plt.legend(loc=1)
plt.xlabel('Mức lương (triệu)')
plt.ylabel('Kinh nghiệm (năm)')
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))

x = np.hstack((np.ones((N, 1)), x))
w = np.array([0., 0.1, 0.1]).reshape(-1, 1)

num = 1000
cost = np.zeros((num, 1))
learning = 0.01

for i in range(num):
    y_per = sigmoid(np.dot(x, w))
    cost[i] = -np.sum(np.multiply(y, np.log(y_per)) + np.multiply(1 - y, np.log(1 - y_per)))

    w = w - learning * np.dot(x.T, y_per - y)
    print(cost[i])
```

Để làm điều này, giả sử quan hệ giữa chiều cao x và cân nặng y là quan hệ tuyến
tính y = w0 + w1x.

```python
from sklearn import ensemble

x=np.array([[145, 150, 152, 154, 156, 160, 162, 164, 165, 168,170,172,174,175,177,178,180,181,182,184]]).T

y =[42,46, 50, 54, 54, 56, 60, 64, 65, 66,67,68,69,70,71,72, 72,73,74,75]

N=len(x)
onesMat = np.ones((1,N))
X=np.concatenate((onesMat,x.T),axis=0)

a=np.dot(X,X.T)
b=np.dot(X,y)

w=np.dot(np.linalg.pinv(a),b)
w0=-1
w1=1

x0=np.linspace(145,184,2)
y0=w0+w1*x0
plt.plot(x0,y0)
plt.plot(x,y,'bo',c='r')
plt.title('Hàm xấp xỉ')
plt.xlabel('Chiều cao (cm)')
plt.ylabel('Cân nặng (kg)')
plt.show()
```
