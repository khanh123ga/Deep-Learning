Đầu tiên tìm tỉ lệ mẩu của viên màu xanh lá bằng cách lấy tổng số viên màu xanh lá/tổng số lượng viên trong mẫu

```python
sum_col = df.sum()
print(sum_col)

Totalkeo=sum_col.sum()-(sum_col.Bag.sum()+sum_col.Weight.sum())
print('Tổng số kẹo',Totalkeo)

sum_green = sum_col.Green.sum()
print('Tổng số kẹo màu xanh',sum_green)
```

ở đây ta biết được Green   222
Totakeo 1713


```python
TLMgr = sum_green/Totalkeo
print(TLMgr)
```

```python
from scipy import stats

confidence_level = 0.95

# Tính độ lệch chuẩn của tỷ lệ mẫu
std = np.sqrt((TLMgr * (1 - TLMgr)) /Totalkeo)

# Tính giá trị Z ứng với độ tin cậy 95%
z_score = stats.norm.ppf(1 - confidence_level / 2)

# Tính khoảng tin cậy 95%
margin_of_error = z_score * std
lower_bound = TLMgr - margin_of_error
upper_bound = TLMgr + margin_of_error

print(f"Khoảng tin cậy 95%: ({lower_bound:.4f}, {upper_bound:.4f})")
```
