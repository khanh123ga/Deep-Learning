Kiểm định tham số/phân phối


 Kiểm định Shapiro-Wilk:



Kiểm định Shapiro-Wilk kiểm tra xem một mẫu dữ liệu có thuộc phân phối chuẩn hay không.
Null hypothesis (H0): Dữ liệu tuân theo phân phối chuẩn.
Alternative hypothesis (H1): Dữ liệu không tuân theo phân phối chuẩn.
Nếu giá trị p (p-value) thu được từ kiểm định Shapiro-Wilk là lớn (lớn hơn một ngưỡng alpha đã chọn, thường là 0.05), bạn không thể bác bỏ giả thuyết null, và dữ liệu có thể được coi là tuân theo phân phối chuẩn.


Kiểm định Kolmogorov-Smirnov:



Kiểm định Kolmogorov-Smirnov kiểm tra xem một mẫu dữ liệu có tuân theo một phân phối chuẩn (hoặc bất kỳ phân phối nào khác đã chỉ định) hay không.
Null hypothesis (H0): Dữ liệu tuân theo phân phối chuẩn (hoặc phân phối đã chỉ định).
Alternative hypothesis (H1): Dữ liệu không tuân theo phân phối chuẩn (hoặc phân phối đã chỉ định).
Tương tự, bạn sẽ kiểm tra giá trị p. Nếu p lớn hơn alpha (thường là 0.05), bạn không thể bác bỏ giả thuyết null, và dữ liệu có thể được coi là tuân theo phân phối chuẩn (hoặc phân phối đã chỉ định).



Dựa vào giá trị p, bạn có thể quyết định xem dữ liệu có tuân theo phân phối chuẩn hay không. Thường thì nếu giá trị p lớn hơn alpha, bạn sẽ không bác bỏ giả thuyết null và kết luận rằng dữ liệu có thể tuân theo phân phối chuẩn.

```python
#Dùng kiểm định theo thu vien shapiro
from scipy.stats import shapiro


statistic, p_value =shapiro(data_cl['fixed acidity'])

# In kết quả
print("Kiểm định Shapiro-Wilk:")
print("Statistic:", statistic)
print("p-value:", p_value)

apha = 0.05
if p_value >apha:
  print("Dữ liệu có thể tuân theo phân phối chuẩn.")
else:
    print("Dữ liệu không tuân theo phân phối chuẩn.")

```

```python
#kiểm định theo thu vien Kolmogorov-Smirnov
from scipy.stats import kstest
#statistic đánh giá mức độ tương tự giữa phân phối thực tế và phân phối thử nghiệm
statistic, p_value = kstest(data_cl['fixed acidity'],'norm')

# In kết quả
print("Kiểm định Shapiro-Wilk:")
print("Statistic:", statistic)
print("p-value:", p_value)

apha = 0.05
if p_value >apha:
  print("Dữ liệu có thể tuân theo phân phối chuẩn.")
else:
    print("Dữ liệu không tuân theo phân phối chuẩn.")
```

```python
def load_data(data):
  y=data['price']
  y=np.array([y]).T
  y=y/10**6

  choose_features=[
      'bedrooms',
      'bathrooms',
      'floors',
      'view',
      'condition',
      'grade',
      'yr_built',
      'sqft_living',
      'sqft_lot',
      'sqft_above',
      'sqft_basement',]
  X= data[choose_features]
  X= np.array(X)
  X=X/10**6
  one = np.ones((X.shape[0],1))

  X=np.hstack((one,X))
  return X,y

```

```python
X,y = load_data(data_train)
```

```python
XtX = X.T @ X

XtX_inv = np.linalg.inv(XtX)

Xty = X.T @ y

theta = XtX_inv @ Xty
theta = theta[: , 0]
```

```python
print(np.linalg.det(XtX))
```

```python
regr = LinearRegression(fit_intercept=False).fit(X,y)
theta_sk = regr.coef_[0]
```

```python
print('theta tính bằng công thức',theta)
print('theta tính bằng thư viện',theta_sk)
```

```python
#hàm mất mát
def loss_mm (X,y,theta):
  #reshape thay đổi hình dạng của mảng
  theta=theta.reshape(-1,1)
  m=len(X)
  # @ phép nhân ma trận với vector
  h=X @ theta
  loss = 1/(2*m)* np.sum((y-h)**2)
  return loss
```

So sánh kết quả của hàm mất mát bằng 2 kết quả


```python
print('KQ tính bằng công thức',loss_mm(X,y,theta))
print('KQ tính bằng thư viện',loss_mm(X,y,theta_sk))
```

```python
def dl(X,y,theta):
  theta = theta.reshape(-1,1)
  m=len(X)
  return -(1/m)*np.sum(X*(y-X @ theta),axis=0)
```

```python
def gradient_descemt():
   m,n = X.shape

   theta = np.random.randn(n)
   iter = 100001
   apha = 0.5

   for i in range(iter):

      theta = theta - apha *dl(X,y,theta)


      loss = loss_mm(X,y,theta)

      if i%20000 ==0:
        print('Iter',i,'loss',loss)

   return theta
```

```python
theta_gd = gradient_descemt()
```

```python
print('KQ tính bằng công thức',loss_mm(X,y,theta_gd))
print('KQ tính bằng thư viện',loss_mm(X,y,theta_sk))
```
